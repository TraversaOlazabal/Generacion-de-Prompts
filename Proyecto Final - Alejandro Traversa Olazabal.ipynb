### Presentación del problema a abordar

El mundo de las inversiones puede parecer complejo y abrumador para personas sin conocimientos previos. La falta de educación financiera y el temor a perder dinero impiden que muchos se animen a invertir, perdiendo la oportunidad de generar ingresos pasivos y asegurar un mejor futuro económico.

En un contexto económico incierto, donde la volatilidad de los mercados afectan directamente la capacidad de ahorro e inversión de las personas, muchas familias y pequeños inversores carecen de acceso a asesoramiento financiero profesional. Esto limita su capacidad para preservar o hacer crecer su capital.

La problemática radica en que:

1.	Las opciones de inversión suelen ser confusas para personas sin conocimientos técnicos.
2.	Existe una barrera de acceso al asesoramiento financiero personalizado, ya sea por costos elevados o falta de disponibilidad en zonas específicas.
3.	Las herramientas digitales existentes no siempre están adaptadas al contexto local ni consideran perfiles de riesgo variados.

La relevancia de desarrollar una solución radica en democratizar el acceso a recomendaciones financieras básicas, ayudando a pequeños inversores a tomar decisiones informadas, mejorar su economía personal y reducir la incertidumbre sobre el futuro de su capital.

### Desarrollo de la propuesta de solución

Descripción de la solución:

El proyecto consiste en desarrollar un asistente virtual basado en inteligencia artificial que permita a los usuarios:

1.	Ingresar sus preferencias de inversión (monto disponible, horizonte temporal, perfil de riesgo).
2.  Recibir un análisis automatizado que incluya:

    • Sugerencias textuales detalladas sobre dónde invertir su dinero según los parámetros ingresados.

    • Una visualización gráfica que anime al potencial inversor y acompañe la recomendación.

El asistente utilizará dos modelos de IA:

Texto-Texto: para generar recomendaciones financieras personalizadas.
    
Texto-Imagen: para generar imágenes que acompañen a la recomendación.

**Prompts utilizados:**

1.	*Para el cotexto (rol de sistema):* "Eres un asesor financiero en Argentina que ofrece recomendaciones personalizadas de instrumentos financieros    basándote en el perfil de riesgo del inversor, el monto de capital disponible y el plazo de inversión."
2.  *Para el modelo Texto-Texto:* Perfil del inversor: {perfil} ; Capital disponible: {capital} USD ; Plazo de inversión: {plazo} meses
¿Qué instrumentos financieros recomendarías para el siguiente caso y por qué?
Por ejemplo, para un inversor con perfil **agresivo**, **$10,000 USD** de capital disponible y un plazo de **6 meses**, una posible recomendación podría ser:
- **Acciones de empresas tecnológicas**: 50% del capital, debido a su potencial de alto rendimiento en el corto plazo.
- **Criptomonedas**: 50% del capital, considerando la alta rentabilidad pero también la volatilidad asociada.
3.	*Para el modelo Texto-Imagen:*
- "success and target"

Luego de varias pruebas estos prompts han sido los que porcentualmente mejores resultados han mostrado por parte del asistente.
Estos prompts serán adaptados dinámicamente a los datos ingresados por el usuario en el sitio web. **(Interactividad en la introducción de Pompts)**

**Detalle de los costos de texto a imagen:**

```python
#Importar bibliotecas de openAI (cargar con conda install openai) y otras necesarias
import openai
import matplotlib.pyplot as plt
import requests
import os
from io import BytesIO
from PIL import Image


#Utilizar la API Key de la página de openAI ( https://platform.openai.com/settings/organization/api-keys )
openai.api_key = ""

# Función para obtener recomendaciones de inversión
def obtener_recomendacion_inversion(perfil, capital, plazo):
    # Contexto del sistema
    contexto = """
    Eres un asesor financiero en Argentina que ofrece recomendaciones personalizadas de instrumentos financieros
    basándote en el perfil de riesgo del inversor, el monto de capital disponible y el plazo de inversión.
    """

    # Creación del prompt (utilizando la técnica de "one-shot prompting")
    prompt = f"""
    Perfil del inversor: {perfil}
    Capital disponible: {capital} USD
    Plazo de inversión: {plazo} meses

    ¿Qué instrumentos financieros recomendarías para el siguiente caso y por qué?
    Por ejemplo, para un inversor con perfil **agresivo**, **$10,000 USD** de capital disponible y un plazo de **6 meses**, una posible recomendación podría ser:
    - **Acciones de empresas tecnológicas**: 50% del capital, debido a su potencial de alto rendimiento en el corto plazo.
    - **Criptomonedas**: 50% del capital, considerando la alta rentabilidad pero también la volatilidad asociada.
    """

    # Llamado a la API de OpenAI
    try:
        respuesta = openai.ChatCompletion.create(
            model="gpt-4o-mini-2024-07-18",
            messages=[
                {"role": "system", "content": contexto},
                {"role": "user", "content": prompt}
            ],
            max_tokens=500  # Ajustar según el nivel de detalle esperado
        )
        
        # Respuesta del modelo
        return respuesta.choices[0].message['content']

    except Exception as e:
        return f"Error al obtener respuesta: {e}"

# Función principal
def main():
    print("¡Bienvenido al Asesor Financiero Virtual!\n")
    
    # Solicitar información al usuario (Interactividad en la introducción de Pompts)
    perfil = input("Por favor, indica tu perfil de inversor (Conservador, Moderado, Agresivo): ").capitalize()
    capital = float(input("¿Cuál es el monto de capital disponible para invertir (en USD)? "))
    plazo = int(input("¿Cuál es el plazo de la inversión (en meses)? "))

    # Obtener recomendación del modelo
    print("\nProcesando tu información...\n")
    recomendacion = obtener_recomendacion_inversion(perfil, capital, plazo)
    
    # Mostrar la recomendación
    print("Recomendación de Inversión:")
    print(recomendacion)

# Ejecutar la función principal si se corre el script
if __name__ == "__main__":
    main()
```

```python
# Prompt de la imagen a crear
imagen_frase = f"success and target"

# Solicitud al asistente (API DALL-E2)
image_response = openai.Image.create(
    prompt=imagen_frase,
    n=1,
    size="512x512"
)

# Obtener y mostrar URL de la imagen
print(image_response['data'][0]['url'])
```

```python
# Descargar la imagen
response = requests.get(image_response['data'][0]['url'])

# Verificar si es posible descargar
if response.status_code == 200:
    # Obtener nombre del archivo
    filename = os.path.basename(image_response['data'][0]['url'])
        # Agregar la extensión ".png"
    filename += ".png"
    
    # Carpeta para guardar la imagen
    directory = "./Imagenes Generadas/"
    
    # Crear la carpeta si esta no existiera
    if not os.path.exists(directory):
        os.makedirs(directory)
    
    # Carpeta y archivo
    filepath = os.path.join(directory, filename)
    
    # Guardar la imagen
    with open(filepath, 'wb') as f:
        f.write(response.content)
        
    print("La imagen ha sido descargada y almacenada correctamente en:", filepath)
    img = Image.open(filepath)
    print(f"")
    display(img)

else:
    print("Hubo un error al descargar la imagen.")
```
